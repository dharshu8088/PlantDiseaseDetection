```python
# --- IMPORTS ---
import os
import zipfile
import random
import shutil
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.ensemble import RandomForestClassifier
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# --- PART 1: WEATHER-BASED DISEASE PREDICTION ---
print("\n=== Weather-Based Disease Prediction ===")

weather_path = 'alover (1).csv'
weather_df = pd.read_csv(weather_path)

selected_features = ['humidity', 'precip', 'dew', 'windspeed', 'temp', 'cloudcover']
weather_features = weather_df[selected_features].copy()

weather_df['disease_label'] = ((weather_df['humidity'] > 70) & 
                               (weather_df['precip'] > 0.3) & 
                               (weather_df['temp'] < 30) & 
                               (weather_df['cloudcover'] > 4)).astype(int)

np.random.seed(42)
flip_mask = np.random.rand(len(weather_df)) < 0.1
weather_df['disease_label'] = np.where(flip_mask, 1 - weather_df['disease_label'], weather_df['disease_label'])
weather_labels = weather_df['disease_label']

sns.countplot(x=weather_labels)
plt.title("Noisy Disease Label Distribution (10% noise)")
plt.xlabel("Label (0: No Disease, 1: Disease)")
plt.ylabel("Count")
plt.show()

scaler = StandardScaler()
weather_features_scaled = scaler.fit_transform(weather_features)

X_train, X_test, y_train, y_test = train_test_split(
    weather_features_scaled, weather_labels, 
    test_size=0.3, stratify=weather_labels, random_state=42
)

model_weather = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    min_samples_leaf=1,
    random_state=42
)
model_weather.fit(X_train, y_train)

y_pred = model_weather.predict(X_test)
accuracy_weather = accuracy_score(y_test, y_pred)
print(f"\n✅ Weather Model Accuracy: {accuracy_weather:.4f}")
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['No Disease', 'Disease'],
            yticklabels=['No Disease', 'Disease'])
plt.title("Weather Model Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

# --- PART 2: CNN IMAGE CLASSIFIER FOR ALOE VERA DISEASE ---
print("\n=== Image-Based CNN Disease Detection ===")

zip_path = 'archive (5).zip'
extract_path = 'aloe_vera_images'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print(f"✅ Extracted to: {extract_path}")

folders = os.listdir(extract_path)
base_data_path = os.path.join(extract_path, folders[0]) if len(folders) == 1 else extract_path
train_dir = os.path.join(extract_path, 'train')
valid_dir = os.path.join(extract_path, 'valid')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(valid_dir, exist_ok=True)

for class_folder in os.listdir(base_data_path):
    class_path = os.path.join(base_data_path, class_folder)
    if os.path.isdir(class_path):
        images = [img for img in os.listdir(class_path) if img.endswith(('.jpg', '.png', '.jpeg'))]
        random.shuffle(images)
        split_idx = int(0.8 * len(images))
        train_imgs = images[:split_idx]
        valid_imgs = images[split_idx:]

        os.makedirs(os.path.join(train_dir, class_folder), exist_ok=True)
        os.makedirs(os.path.join(valid_dir, class_folder), exist_ok=True)

        for img in train_imgs:
            shutil.copy(os.path.join(class_path, img), os.path.join(train_dir, class_folder, img))
        for img in valid_imgs:
            shutil.copy(os.path.join(class_path, img), os.path.join(valid_dir, class_folder, img))

print("✅ Image dataset split into train and validation folders.")

img_size = (150, 150)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

valid_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

valid_generator = valid_datagen.flow_from_directory(
    valid_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

model_cnn = models.Sequential([
    layers.Input(shape=(150, 150, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model_cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model_cnn.summary()

history = model_cnn.fit(
    train_generator,
    epochs=12,
    validation_data=valid_generator
)

y_true = valid_generator.classes
y_probs = model_cnn.predict(valid_generator)
y_pred = np.argmax(y_probs, axis=1)

acc_cnn = accuracy_score(y_true, y_pred)
print(f"\n✅ CNN Image Classifier Accuracy: {acc_cnn:.4f}")

label_map = {v: k for k, v in valid_generator.class_indices.items()}
target_names = [label_map[i] for i in np.unique(y_true)]

print("\n📊 CNN Classification Report:\n", classification_report(
    y_true, y_pred,
    target_names=target_names
))

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names,
            yticklabels=target_names)
plt.title("CNN Image Classifier Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("CNN Accuracy over Epochs")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("CNN Loss over Epochs")
plt.legend()

plt.tight_layout()
plt.show()

# --- PART 3: LIVE CAMERA PREDICTION (AUTO EXIT ON DETECTION) ---
def predict_from_camera_and_exit_on_detection(model, label_map, confidence_threshold=0.7):
    cap = cv2.VideoCapture(0)
    detected = False

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        img = cv2.resize(frame, (150, 150)) / 255.0
        img = img.reshape(1, 150, 150, 3)

        preds = model.predict(img)
        pred_class = np.argmax(preds)
        confidence = np.max(preds)

        if confidence > confidence_threshold:
            label = label_map[pred_class]
            message = f"Aloe Vera Detected: {label} ({confidence:.2f}) ✅"
            detected = True
        else:
            message = "⚠️ Not Aloe Vera - Try Again"

        cv2.putText(frame, message, (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8,
                    (0, 255, 0) if detected else (0, 0, 255), 2)

        cv2.imshow("Live Aloe Vera Classifier", frame)

        if detected:
            print(f"\n✅ Aloe Vera detected: {label} with {confidence:.2f} confidence.")
            break

        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("❌ User exited manually.")
            break

    cap.release()
    cv2.destroyAllWindows()

# --- RUN LIVE CAMERA ---
print("\n=== Starting Live Camera Aloe Vera Detection ===")
predict_from_camera_and_exit_on_detection(model_cnn, label_map)
```

```python
import gradio as gr
from PIL import Image
import numpy as np

# --- CNN-based image prediction function ---
def predict_from_image(img):
    if img is None:
        return "⚠️ No image received."
    
    img = img.convert("RGB")
    img = img.resize((150, 150))
    img_array = np.array(img) / 255.0
    img_array = img_array.reshape(1, 150, 150, 3)

    preds = model_cnn.predict(img_array)
    pred_class = np.argmax(preds)
    confidence = float(np.max(preds))

    label = label_map[pred_class]
    return f"🖼️ Image Prediction: {label} ({confidence:.2f} confidence)"

# --- Weather-based prediction function ---
def predict_from_weather(humidity, precip, dew, windspeed, temp, cloudcover):
    input_data = np.array([[humidity, precip, dew, windspeed, temp, cloudcover]])
    input_scaled = scaler.transform(input_data)
    prediction = model_weather.predict(input_scaled)[0]
    label = "Disease Detected" if prediction == 1 else "No Disease"
    return f"🌦️ Weather Prediction: {label}"

# --- Gradio Interface ---
with gr.Blocks(title="Aloe Vera Disease Predictor") as demo:
    gr.Markdown("## 🌿 Aloe Vera Disease Detection System")

    with gr.Tab("📷 Image Classifier"):
        gr.Markdown("Upload or capture an Aloe Vera leaf image to detect disease.")
        image_input = gr.Image(type="pil", label="Upload or Webcam Image")
        image_output = gr.Textbox(label="Prediction Result")
        image_button = gr.Button("Predict from Image")
        image_button.click(fn=predict_from_image, inputs=image_input, outputs=image_output)

    with gr.Tab("🌤️ Weather Classifier"):
        gr.Markdown("Enter weather parameters to predict Aloe Vera disease.")
        humidity = gr.Slider(0, 100, label="Humidity (%)")
        precip = gr.Slider(0.0, 5.0, step=0.1, label="Precipitation (mm)")
        dew = gr.Slider(0.0, 30.0, step=0.1, label="Dew Point (°C)")
        windspeed = gr.Slider(0.0, 30.0, step=0.1, label="Wind Speed (km/h)")
        temp = gr.Slider(0.0, 50.0, step=0.5, label="Temperature (°C)")
        cloudcover = gr.Slider(0, 9, step=1, label="Cloud Cover (oktas)")

        weather_output = gr.Textbox(label="Prediction Result")
        weather_button = gr.Button("Predict from Weather")
        weather_button.click(
            fn=predict_from_weather,
            inputs=[humidity, precip, dew, windspeed, temp, cloudcover],
            outputs=weather_output
        )

demo.launch(inline=True)
```
